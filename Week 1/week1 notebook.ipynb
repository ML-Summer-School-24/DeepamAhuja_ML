```python
import pandas as pd
```

DATAFRAME WITH DETAILS OF 10 STUDENTS

```python
data = pd.read_csv('student.csv')
data
```

Create a new column with total marks

```python
data['Total Marks'] = data['Marks 1'] + data['Marks 2'] + data['Marks 3']
data 
```

FIND STUDENT WITH LOWEST MARKS IN MARKS 1

```python
data['Marks 1'].min()
```

FIND STUDENT WITH HIGHEST MARKS IN MARKS 2

```python
data['Marks 2'].max()
```

FIND AVERAGE OF MARKS 3

```python
data['Marks 3'].mean()
```

STUDENT WITH HIGHEST AVERAGE

```python
data['Average'] = data['Total Marks'] / 3
data.sort_values(by='Average', ascending=False)
```

NUMBER OF STUDENT WHO FAILED MARKS2 (<80)

```python
dataframe = data[data['Marks 2']<80]
dataframe
```
